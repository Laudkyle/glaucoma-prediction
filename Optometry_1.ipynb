```python
# Checking for the presence of categorical columns

def is_categorical(df):
    categorical_columns =[]
    cols = df.columns.values
    for col in cols:
        if df[col].dtype == "object":
            categorical_columns.append(col)
    return categorical_columns

print(is_categorical(df_main))
```

```python
# checking for outliers

def plot_all_columns_for_outliers(df):
    """
    Plots box plots for all columns in the dataframe to check for outliers.
    """
    num_columns = len(df.columns)
    plt.figure(figsize=(20, num_columns * 5))
    
    for i, column in enumerate(df.columns, 1):
        plt.subplot(num_columns, 1, i)
        sns.boxplot(x=df[column])
        plt.title(f'Box Plot for {column}')
    
    plt.tight_layout()
    plt.show()

plot_all_columns_for_outliers(df_main)
```

```python
def identify_columns_with_outliers(df):
    """
    Identifies columns in the dataframe that contain outliers using the IQR method.

    """
    columns_with_outliers = []

    for column in df.columns:
        if pd.api.types.is_numeric_dtype(df[column]):
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            
            if ((df[column] < lower_bound) | (df[column] > upper_bound)).any():
                columns_with_outliers.append(column)

    return columns_with_outliers


columns_with_outliers = identify_columns_with_outliers(df_main)
```

```python
def log_transform_replace_outliers(df, columns):
    """
    Applies log transformation to specified columns in the dataframe and replaces the original columns with the transformed values.
    Returns:
    pd.DataFrame: The dataframe with specified columns log-transformed.
    """
    df_transformed = df.copy()
    for column in columns:
        df_transformed[column] = df_transformed[column].apply(lambda x: np.log1p(x) if x > 0 else x)
    return df_transformed


df_main_transformed = log_transform_replace_outliers(df_main, columns_with_outliers)
```

```python
# Define custom scoring functions
def accuracy(y_true, y_pred):
    return accuracy_score(y_true, y_pred)

def sensitivity(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    return cm[1, 1] / (cm[1, 0] + cm[1, 1])

def specificity(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    return cm[0, 0] / (cm[0, 0] + cm[0, 1])

# Define the function to perform grid search
def perform_grid_search(X_train, y_train, models, balance=False):
    results = []
    if balance:
        smote = SMOTE()  # Initialize SMOTE
        X_train, y_train = smote.fit_resample(X_train, y_train)

    for model_name, model_info in models.items():
        pipe = Pipeline([
            ('model', model_info['model'])
        ])
        
        grid_search = GridSearchCV(estimator=pipe, 
                                   param_grid=model_info['params'], 
                                   scoring='accuracy',  # Use accuracy or any other metric
                                   refit=True,  # Refit on the best score
                                   cv=5,  # Number of cross-validation folds
                                   n_jobs=-1, 
                                   verbose=2)  
        
        grid_search.fit(X_train, y_train)
        
        # Get the best model from grid search
        best_model = grid_search.best_estimator_
        
        results.append({
            'model': model_name,
            'best_params': grid_search.best_params_,
            'best_accuracy': grid_search.best_score_,
            'best_estimator': best_model,
            'cv_results_': grid_search.cv_results_  # Include all CV results for detailed analysis
        })

    return results

# Function to calculate AUROC separately
def calculate_auroc(best_model, X_test, y_test):
    # Handle models that don't have predict_proba or decision_function (e.g., RidgeClassifier)
    if hasattr(best_model, 'predict_proba'):
        y_probs = best_model.predict_proba(X_test)[:, 1]  # Probabilities for the positive class
    elif hasattr(best_model, 'decision_function'):
        y_probs = best_model.decision_function(X_test)  # Decision function scores
    else:
        raise ValueError(f"Model {best_model} does not support probability estimates or decision function.")

    auroc = roc_auc_score(y_test, y_probs)
    return auroc

# Function to calculate sensitivity and specificity
def calculate_sensitivity_specificity(best_model, X_test, y_test):
    y_pred = best_model.predict(X_test)
    sens = sensitivity(y_test, y_pred)
    spec = specificity(y_test, y_pred)
    return sens, spec

# Define a function to print the results
def print_results(results, X_test, y_test):
    for result in results:
        print(f"Model: {result['model']}")
        print(f"Best Parameters: {result['best_params']}")
        print(f"Best Accuracy: {result['best_accuracy']:.4f}")
        
        # Calculate AUROC on the test set using the best model
        best_auroc = calculate_auroc(result['best_estimator'], X_test, y_test)
        print(f"Test AUROC: {best_auroc:.4f}")
        
        # Calculate Sensitivity and Specificity on the test set using the best model
        best_sensitivity, best_specificity = calculate_sensitivity_specificity(result['best_estimator'], X_test, y_test)
        print(f"Test Sensitivity: {best_sensitivity:.4f}")
        print(f"Test Specificity: {best_specificity:.4f}")
        
        print("-" * 80)
```

## Getting the weights of the columns

```python
def extract_feature_weights(model, feature_names):
    """
    Extracts feature weights or importances from the trained model.
    """
    if hasattr(model, 'coef_'):
        # Logistic Regression or Linear SVM
        weights = model.coef_.flatten()
    elif hasattr(model, 'feature_importances_'):
        # Tree-based models
        weights = model.feature_importances_
    else:
        # Models without feature weights
        return pd.DataFrame(columns=['Feature', 'Weight'])

    return pd.DataFrame({
        'Feature': feature_names,
        'Weight': weights
    }).sort_values(by='Weight', ascending=False)
```

```python
def perform_grid_search_weights(X_train, y_train, models):
    """
    Performs GridSearchCV for multiple models and returns the results with weights included.
    """
    results = []
    top_features_set = set()
    feature_names = X_train.columns.tolist() if isinstance(X_train, pd.DataFrame) else [f'Feature_{i}' for i in range(X_train.shape[1])]

    for model_name, model_info in models.items():
        pipe = Pipeline([
            ('model', model_info['model'])
        ])
        
        grid_search = GridSearchCV(estimator=pipe, 
                                   param_grid=model_info['params'], 
                                   scoring='accuracy',  # You can use different scoring metrics
                                   cv=5,  # Number of cross-validation folds
                                   n_jobs=-1,  # Use all available cores
                                   verbose=2)  # Verbosity level
        
        grid_search.fit(X_train, y_train)
        
        best_model = grid_search.best_estimator_.named_steps['model']
        feature_weights = extract_feature_weights(best_model, feature_names)
        top_features = get_top_features(feature_weights)

        # Add top features to the set
        top_features_set.update(top_features)


        results.append({
            'model': model_name,
            'best_params': grid_search.best_params_,
            'best_score': grid_search.best_score_,
            'best_estimator': grid_search.best_estimator_,
            'feature_weights': feature_weights,
        })

    return results, top_features_set


def print_results_with_weights(results):
    for result in results:
        print(f"Model: {result['model']}")
        print(f"Best Parameters: {result['best_params']}")
        print(f"Best Score: {result['best_score']:.4f}")
        print("Feature Weights:")
        if not result['feature_weights'].empty:
            print(result['feature_weights'])
        else:
            print("No feature weights available for this model.")
        print("-" * 50)

def get_top_features(feature_weights_df, top_n=3):
    return feature_weights_df.head(top_n)['Feature'].tolist()
```

```python
def fit_group_with_weights(group):
    X = df_main_standardized[group]
    y = df_main['Dx_Hos']
    
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=101)
    results, top_features = perform_grid_search_weights(X_train, y_train, models)
    return results, top_features
```

```python

from sklearn import svm
from itertools import combinations


# Define the SVM classifier
clf = svm.SVC(kernel='linear')

# Get the number of features
num_features = X.shape[1]

# Create a list of all pairs of features
feature_pairs = list(combinations(range(num_features), 2))

# Set up the figure
plt.figure(figsize=(15, 15))

# Iterate over each pair of features
for i, (feat1, feat2) in enumerate(feature_pairs):
    # Select the two features for the current pair using Pandas indexing
    X_pair = X.iloc[:, [feat1, feat2]]
    
    # Fit the SVM model on the current pair
    clf.fit(X_pair, y)
    
    # Define the meshgrid for the plot
    h = .02  # Step size in the mesh
    x_min, x_max = X_pair.iloc[:, 0].min() - 1, X_pair.iloc[:, 0].max() + 1
    y_min, y_max = X_pair.iloc[:, 1].min() - 1, X_pair.iloc[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    # Get decision function for the grid points
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    # Create a subplot for each pair
    plt.subplot(len(feature_pairs)//2 + 1, 2, i + 1)
    plt.contourf(xx, yy, Z > 0, alpha=0.3, cmap=plt.cm.Paired)
    plt.scatter(X_pair.iloc[:, 0], X_pair.iloc[:, 1], c='red', edgecolors='k', cmap=plt.cm.Paired)
    plt.xlabel(f'GLV {feat1 + 1}')
    plt.ylabel(f'FLV {feat2 + 1}')
    plt.title(f'SVM Decision Boundary: Feature {feat1 + 1} vs Feature {feat2 + 1}')

plt.tight_layout()
plt.show()
```
